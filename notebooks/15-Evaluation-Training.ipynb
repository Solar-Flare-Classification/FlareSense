```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("data/raw/results/results_training.csv")
df
```

```python
df.isna().sum()
```

```python
df = df.rename(columns={"lr": "Learning Rate", "weight_decay": "L2 Regularisierung", "val_precision": "Validation Precision", "val_recall": "Validation Recall"})
df
```

```python
df = df[df["Validation Recall"] != 0]
df
```

```python
df = df[df["use_data_augmentation"] == True]
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.groupby("model")["Validation Precision"].count()
```

```python
df.groupby("model")[["Validation Precision", "Validation Recall"]].agg(["count", "min", "max", "median"])
```

```python
df.groupby("model")["Validation Recall"].agg(["max"]).plot(kind="bar", legend=None, figsize=(10, 4), rot=0)
plt.title("Maximale Validation Recall pro Modell")
plt.ylim((0.6, 1))
plt.show()
```

```python
df.groupby("model")["Validation Precision"].agg(["max"]).plot(kind="bar", legend=None, figsize=(10, 4), rot=0)
plt.title("Maximale Validation Precision pro Modell")
plt.ylim((0.6, 1))
plt.show()
```

```python
df_par_cord = df.copy()
df_par_cord["prov Validation Recall"] = df_par_cord["Validation Recall"]
df_par_cord
```

```python
df[df["model"] == "ResNet50"]
```

```python
df_ResNet50 = df[df["model"] == "ResNet50"]
df_ResNet18 = df[df["model"] == "ResNet18"]
df_CustomCNN = df[df["model"] == "CustomCNN"]
df_CustomCNN2 = df[df["model"] == "CustomCNN2"]

df_ResNet50.columns
```

```python
plt.scatter(df_ResNet50["Validation Recall"], df_ResNet50["Validation Precision"])
plt.xlabel("Validation Recall")
plt.ylabel("Validation Precision")
plt.title("Precision Recall ResNet50")
plt.show()
```

```python
plt.scatter(df_ResNet18["Validation Recall"], df_ResNet18["Validation Precision"])
plt.xlabel("Validation Recall")
plt.ylabel("Validation Precision")
plt.title("Precision Recall ResNet18")
plt.show()
```

```python
plt.scatter(df_CustomCNN2["Validation Recall"], df_CustomCNN2["Validation Precision"])
plt.xlabel("Validation Recall")
plt.ylabel("Validation Precision")
plt.title("Precision Recall CustomCNN2")
plt.show()
```

```python
plt.scatter(df["Validation Recall"], df["Validation Precision"])
plt.xlabel("Validation Recall")
plt.ylabel("Validation Precision")
plt.title("Precision Recall")
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=200)

plt.scatter(df_CustomCNN["Validation Recall"], df_CustomCNN["Validation Precision"], color = "black", label="CustomCNN", s=10)
plt.scatter(df_CustomCNN2["Validation Recall"], df_CustomCNN2["Validation Precision"], color = "green", label="CustomCNN2", s=10)
plt.scatter(df_ResNet18["Validation Recall"], df_ResNet18["Validation Precision"], color = "red", label="ResNet18", s=10)
plt.scatter(df_ResNet50["Validation Recall"], df_ResNet50["Validation Precision"], color = "blue", label="ResNet50", s=10)

plt.xlabel("Validation Recall")
plt.ylabel("Validation Precision")
plt.title("Vergleich der Modell Metriken")
plt.ylim(0,1)
plt.xlim(0,1)
plt.axvline(0.9, color='orange', linestyle = 'dashed')
plt.legend()
plt.show()
```
