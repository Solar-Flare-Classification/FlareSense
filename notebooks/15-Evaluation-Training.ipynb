```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("data/raw/results/results_training.csv")
df
```

```python
df.isna().sum()
```

```python
# df["Duration"] = df["Duration"].dropna()
df = df.dropna(subset = ['Duration'])
df.isna().sum()
```

```python
df
```

```python
df = df.rename(columns={"lr": "Learning Rate", "weight_decay": "L2 Reg", "val_precision": "Val Precision", "val_recall": "Val Recall"})
df
```

```python
df = df[df["use_data_augmentation"] == True]
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df
```

```python
df.groupby("model")["Val Precision"].count()
```

```python
df.groupby("model")[["Val Precision", "Val Recall"]].agg(["count", "min", "max", "median"])
```

```python
df.groupby("model")["Val Recall"].agg(["max"]).plot(kind="bar", legend=None)
plt.title("Max Val Recall pro Modell")
plt.ylim((0.6, 1))
plt.show()
```

```python
df.groupby("model")["Val Precision"].agg(["max"]).plot(kind="bar", legend=None)
plt.title("Maximale Val Precision pro Modell")
plt.ylim((0.6, 1))
plt.show()
```

```python
df_par_cord = df.copy()
df_par_cord["prov Val Recall"] = df_par_cord["Val Recall"]
df_par_cord
```

```python
pd.plotting.parallel_coordinates(df_par_cord[df_par_cord["model"] == "ResNet50"][["Learning Rate", "L2 Reg", "Val Precision", "Val Recall", "prov Val Recall"]], "prov Val Recall")
```
